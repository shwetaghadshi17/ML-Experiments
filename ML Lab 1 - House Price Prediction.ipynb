```python
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'],axis=1) #dropping the target class column 
y = data['median_house_value'] #target class column stored here
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) #0.2 is the % of data we are using for testing..here 20%
```

```python
train_data = x_train.join(y_train)
```

```python
from sklearn.linear_model import LinearRegression

x_train,y_train = train_data.drop(['median_house_value'],axis = 1),train_data['median_house_value']

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
x_test, y_test = test_data.drop(['median_house_value'],axis = 1),test_data['median_house_value']
```

```python
reg.score(x_test,y_test)
```
